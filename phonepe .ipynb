### 4.1 - Ratio of users to population by state

```python
# 1 - merge the State_Txn_and_Users dataset with the District_Demographics dataset to calculate the ratio of registration users to the
#     population of each state. display the result in a tabular format.

# merge the datasets
merged_data = pd.merge(State_Txn_and_Users, District_Demographics, on='State')
# merged_data

state_population =  merged_data.groupby('State').agg(Total_Registered_Users=('Registered Users', 'sum'),Total_Population=('Population', 'sum')).reset_index()
# state_population.head()


# ratio_state_population = state_population['Total_Registered_Users'] /  state_population['Total_Population']
# ratio_state_population

ratio_state_population = state_population.assign(Ratio_registration_Users_to_Population = state_population['Total_Registered_Users'] / state_population['Total_Population'])
ratio_state_population.head()
# ratio_state_population.shape
```

### 4.3 - Average transaction amount per user

```python
# 1 merge relevant datasets to calculate the average transaction amount per user for each state.

# Calculate total transaction amount and total registered users per state
state_data = State_Txn_and_Users.groupby('State').agg(total_transaction_amount=('Amount (INR)', 'sum'),total_users=('Registered Users', 'sum')).reset_index()

# average transaction amount per user for each state
state_data['avg_transaction_per_user'] = state_data['total_transaction_amount'] / state_data['total_users']

# Display the result
print(state_data[['State', 'avg_transaction_per_user']])
```

### 4.4 - Device brand usage ratio

```python
# 1 - merge the state_device dateset with state_txn_and_user dataset

merge_datastes = pd.merge(State_Txn_and_Users,State_DeviceData, on=['State', 'Year', 'Quarter'], how='inner')
merge_datastes.head(1)
```

```python
# 2 - calculate the ratio of users using each device brand to the total number of registered users in each state.

ratio_device_brand_usage = merge_datastes.groupby(['State', 'Brand']).agg(Total_Registered_Users=('Registered Users_y', 'sum')).reset_index()
# ratio_device_brand_usage.head()

ratio_device_brand_usage['Ratio_Device_Brand_Usage'] = ratio_device_brand_usage['Total_Registered_Users'] / ratio_device_brand_usage['Total_Registered_Users'].sum()
ratio_device_brand_usage.head()
```

# Task - 5 = Data visualization

### 5.1 = PLot the total transaction and amount over time for a selected state

```python
# 1 - create a line plot showing the total number of transaction and the total transaction amount over time(years and quarters) for a selected state.

selected_state = 'Andaman & Nicobar Islands'
state_data = State_Txn_and_Users[State_Txn_and_Users['State'] == selected_state]

state_data['Time'] = state_data['Year'].astype(str) + " Q" + state_data['Quarter'].astype(str)
state_data = state_data.sort_values(['Year', 'Quarter'])
state_data.head()

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(state_data['Time'], state_data['Transactions'], color='blue', marker='o', label='Total Transactions')
ax1.set_xlabel('Time (Year & Quarter)')
ax1.set_ylabel('Total Transactions', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for transaction amount
ax2 = ax1.twinx()
ax2.plot(state_data['Time'], state_data['Amount (INR)'], color='red', marker='x', label='Total Transaction Amount (INR)')
ax2.set_ylabel('Total Transaction Amount (INR)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title(f'Total Transactions and Transaction Amount Over Time ({selected_state})')
plt.xticks(rotation=45)
fig.tight_layout()

plt.show()
```
